```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_values = [1400,600,300,410,250]
exp_labels = ['Home Rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other Utilities']
```

```python
plt.pie(exp_values, labels=exp_labels, autopct='%0.2f%%', explode=[0,0.1,0,0,0], startangle=30)
plt.axis('equal')
plt.show() #In order to hide text information
```
